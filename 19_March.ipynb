Q1. What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its
application.

Solution:

Min-Max scaling is a normalization technique that scales the data in a dataset to a specific range using each feature’s minimum and maximum value. It is used in data preprocessing to scale the values of features in a dataset to a fixed range of [0, 1]. This scaling is done feature-wise in an independent way. Min-Max scaling is useful when we want to compare different features that have different units or scales.

Q2. What is the Unit Vector technique in feature scaling, and how does it differ from Min-Max scaling?
Provide an example to illustrate its application.

Solution:

The Unit Vector technique in feature scaling is a normalization technique that scales the data in a dataset by considering the whole feature vector to be of unit length. This scaling is done feature-wise in an independent way. The Unit Vector technique produces values of range [0, 1]. When dealing with features with hard boundaries, this technique is quite useful. For example, when dealing with image data, the colors can range from only 0 to 255.

The formula for Unit Vector scaling is:

x_scaled = x / ||x||

where x is the original value of the feature and ||x|| is the Euclidean length of the feature vector.

Q3. What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an
example to illustrate its application.

Solution

PCA (Principal Component Analysis) is a statistical technique used for dimensionality reduction in machine learning. It is an unsupervised learning algorithm that converts the observations of correlated features into a set of linearly uncorrelated features with the help of orthogonal transformation. These new transformed features are called the Principal Components. PCA is one of the popular tools that is used for exploratory data analysis and predictive modeling. It is a technique to draw strong patterns from the given dataset by reducing the variances.

```python
from sklearn.decomposition import PCA
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("iris")
```

```python
df
```

```python
X = df.iloc[: , :-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=3)
```

```python
 pca.fit_transform(df_scaled)
```

```python
import numpy as np
eigenvalues = pca.explained_variance_
prop_var = eigenvalues / np.sum(eigenvalues)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,4))
plt.plot(np.arange(1, len(prop_var)+1), 
                   prop_var, marker='o')
plt.xlabel('Principal Component',
           size = 10)
plt.ylabel('Proportion of Variance Explained',
           size = 10)
plt.title('Figure 1: Scree Plot for Proportion of Variance Explained',
          size = 10)
plt.grid(True)
```

Q4. What is the relationship between PCA and Feature Extraction, and how can PCA be used for Feature
Extraction? Provide an example to illustrate this concept.

Solution

PCA can be used for feature extraction. Feature extraction is the process of selecting a subset of relevant features for use in model construction. It is a dimensionality reduction technique that reduces the number of features in a dataset while retaining as much information as possible.

PCA can be used for feature extraction by transforming the original features into a new set of features that are linearly uncorrelated. These new features are called the Principal Components. The Principal Components are ordered by their importance, with the first Principal Component having the highest variance.

PCA can be used to extract the most important features from a dataset and reduce its dimensionality. This can help improve the performance of machine learning algorithms by reducing overfitting and improving generalization.

features, such as company financial data and market trends. Explain how you would use PCA to reduce the

Solution

Principal Component Analysis (PCA) is a technique used to reduce the dimensionality of datasets, increasing interpretability while minimizing information loss. It does this by creating new uncorrelated variables that successively maximize variance.

Here’s how you could use PCA in your stock price prediction project:

**Standardize the Data**: PCA is affected by scale, so you need to standardize the features in your dataset so that they have a mean of 0 and variance of 1.


from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data_rescaled = scaler.fit_transform(data)

**Apply PCA**: You can use PCA from sklearn.decomposition to apply principal component analysis on your dataset. The number of components chosen determines the new dimensionality of your dataset.

from sklearn.decomposition import PCA

pca = PCA(n_components = num_components)

pca.fit(data_rescaled)

reduced_data = pca.transform(data_rescaled)

Remember, while PCA can help to deal with high dimensionality, it also makes your features less interpretable. So, it’s a trade-off between interpretability and reducing dimensionality. Q7. For a dataset containing the following values: [1, 5, 10, 15, 20], perform Min-Max scaling to transform the
values to a range of -1 to 1.

Solution:

Min-Max scaling is a simple method to scale your data between a specified range. The formula for Min-Max scaling is:

$$ X_{\text{scaled}} = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}} \times (\text{Max} - \text{Min}) + \text{Min} $$

In your case, the dataset is [1, 5, 10, 15, 20], and you want to scale it to the range of -1 to 1. So, Min = -1 and Max = 1.

Let's apply the formula to each value in the dataset:

- For X = 1:
$$ X_{\text{scaled}} = \frac{1 - 1}{20 - 1} \times (1 - (-1)) - 1 = -1 $$

- For X = 5:
$$ X_{\text{scaled}} = \frac{5 - 1}{20 - 1} \times (1 - (-1)) - 1 = -0.5 $$

- For X = 10:
$$ X_{\text{scaled}} = \frac{10 - 1}{20 - 1} \times (1 - (-1)) - 1 = 0 $$

- For X = 15:
$$ X_{\text{scaled}} = \frac{15 - 1}{20 - 1} \times (1 - (-1)) - 1 = 0.5 $$

- For X = 20:
$$ X_{\text{scaled}} = \frac{20 - 1}{20 - 1} \times (1 - (-1)) - 1 = 1 $$


Q8. For a dataset containing the following features: [height, weight, age, gender, blood pressure], perform
Feature Extraction using PCA. How many principal components would you choose to retain, and why?

Solution:

The number of principal components to retain depends on the specific requirements of your analysis and the structure of your data. Here are some common strategies:

Variance Explained: Choose the number of components such that a certain percentage of the total variance is retained. For example, you might want to retain enough components to explain 95% of the total variance.

Scree Plot: You can plot the explained variance of each component (in descending order) and look for an “elbow” or inflection point in the plot. The number of components before this point could be a good choice.

Kaiser’s Rule: According to Kaiser’s rule, all the components with eigenvalues greater than 1 should be kept for statistical inference.

Purpose of Analysis: If the purpose is to visualize the data, you might choose to retain only two or three components for a two-dimensional or three-dimensional plot
