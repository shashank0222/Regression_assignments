#  separate independent and dependent features

```python
X = df.drop("Outcome" , axis=1)
y = df["Outcome"]
```

```python
X
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test , y_train , y_test = train_test_split(X,y,test_size=0.33 , random_state=42)
```

```python
X_train.shape , X_test.shape
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_train_scaled
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled
```

#  Pre pruning

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier()
```

```python
#  Hyper parametr tuning
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {
    'criterion' : ['ginni' , 'entropy' , 'log_loss'],
    'splitter' : ['best' , 'random'],
    'max_depth' : [1,2,3,4,5,6],
    'max_features' : ['auto' , 'sqrt' , 'log2'],
    'random_state' : [10,42,22]
    
}
```

```python
clf = GridSearchCV(classifier ,param_grid=parameters , cv=4 ,scoring='accuracy')
```

```python
clf.fit(X_train_scaled , y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
classifier = DecisionTreeClassifier(criterion= 'entropy',max_depth= 5,max_features = 'log2',random_state =22,splitter = 'best')
```

```python
classifier.fit(X_train_scaled,y_train)
```

```python
plt.figure(figsize=(12,10))
from sklearn import tree
tree.plot_tree(classifier , filled = True)
plt.show()
```

```python
y_pred = classifier.predict(X_test_scaled)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
conf_matrix = confusion_matrix(y_test,y_pred)
```

```python
conf_matrix
```

```python
true_positive = conf_matrix[0][0]
false_positive = conf_matrix[0][1]
false_negative = conf_matrix[1][0]
true_negative = conf_matrix[1][1]
```

```python
Accuracy = accuracy_score(y_test,y_pred)
```

```python
Accuracy
```

```python
precision = true_positive / (true_positive + false_positive)
```

```python
precision
```

```python
recall = true_positive / (true_positive + false_negative)
```

```python
recall  # best evaluation metric
```

```python
f1_score = (2 * precision * recall) / (precision + recall)
```

```python
f1_score
```
