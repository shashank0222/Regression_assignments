Q1. What is the relationship between polynomial functions and kernel functions in machine learning
algorithms?

Solution :

Polynomial kernel is a kernel function commonly used with support vector machines (SVMs) and other kernelized models in machine learning. It represents the similarity of vectors in a feature space over polynomials of the original variables, allowing learning of non-linear models.

In a polynomial kernel for SVM, the data is mapped into a higher-dimensional space using a polynomial function. The polynomial kernel is often used in SVM classification problems where the data is not linearly separable.

Q3. How does increasing the value of epsilon affect the number of support vectors in SVR?

Solution :

In support vector regression (SVR), the value of epsilon determines the width of the tube; a smaller value indicates a lower tolerance for error and also affects the number of support vectors and, consequently, the solution sparsity.

If epsilon is decreased, the boundary of the tube is shifted inward. Therefore, more datapoints are around the boundary, which indicates more support vectors. Similarly, increasing epsilon will result in fewer points around the boundary.

Q4. How does the choice of kernel function, C parameter, epsilon parameter, and gamma parameter
affect the performance of Support Vector Regression (SVR)? Can you explain how each parameter works
and provide examples of when you might want to increase or decrease its value?

Solution :

The choice of kernel function, C parameter, epsilon parameter, and gamma parameter can affect the performance of Support Vector Regression (SVR).

* Kernel function: The kernel function is used to transform the training set of data so that a non-linear decision surface is able to transform to a linear equation in a higher number of dimension spaces. The choice of kernel function depends on the data and the problem at hand. Some common kernel functions include linear, polynomial, radial basis function (RBF), sigmoid, etc.

* C parameter: The C parameter trades off correct classification of training examples against maximization of the decision function’s margin. For larger values of C, a smaller margin will be accepted if the decision function is better at classifying all training points correctly. A lower value of C will encourage a larger margin, therefore a simpler decision function, at the cost of training accuracy .

* Epsilon parameter: The epsilon parameter determines the width of the tube; a smaller value indicates a lower tolerance for error and also affects the number of support vectors and, consequently, the solution sparsity .

* Gamma parameter: The gamma parameter defines how far the influence of a single training example reaches. A low value means ‘far’ and high value means ‘close’. The gamma parameters can be seen as the inverse of the radius of influence of samples selected by the model as support vectors .

Here are some examples when you might want to increase or decrease each parameter:

* Kernel function: If you have linearly separable data, you might want to use a linear kernel function. If you have non-linearly separable data, you might want to use a polynomial or RBF kernel function .

* C parameter: If you have noisy data or overlapping classes, you might want to use a smaller value for C. If you have clean data or well-separated classes, you might want to use a larger value for C .

* Epsilon parameter: If you have noisy data or overlapping classes, you might want to use a larger value for epsilon. If you have clean data or well-separated classes, you might want to use a smaller value for epsilon .

* Gamma parameter: If your model is overfitting (i.e., has high variance), you might want to decrease gamma. If your model is underfitting (i.e., has high bias), you might want to increase gamma .

# feature selection

```python
def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold :
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train , 0.80)
```

```python
corr_features
```

```python
#  drop features when corrleation is greater than 0.80
X_train.drop(corr_features , axis = 1 , inplace=True)
X_test.drop(corr_features , axis = 1 , inplace=True)
X_train.shape , X_test.shape
```
